2. Resource-2: https://en.wikipedia.org/wiki/Dyadic_transformation

```python
def func(x):
    return ((2*x)%1)
```

```python
def plot_system(x0, n, ax=None):
    t = np.linspace(0, 1)
    ax.plot(t, func(t), lw=2)
    ax.plot([0, 1], [0, 1], lw=2)
    x = x0
    for i in range(n):
        y = func(x)
        ax.plot([x, x], [x, y], 'k', lw=1)
        ax.plot([x, y], [y, y], 'k', lw=1)
        ax.plot([x], [y])
        x = y
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_title("x0=0.1")


fig, (ax1) = plt.subplots(1)
plot_system(.1, 10, ax=ax1)
```

### Behavior: 
No, it does not have any bifurcating steady states, the cycle's starting and ending point never seems to meet.

```python
def Int(f, a, b, N):
    h = (b-a)/N
    area=0
    x=a
    for i in range (1, N):
        area += f(x)*h
        x=x+h
    return area

a0=int(input())
a1=int(input())
a2=int(input())

a=int(input())
b=int(input())

print("Area is: ", Int(lambda y: a2*(y**2) + a1*y + a0 , a, b, 10000))
```
